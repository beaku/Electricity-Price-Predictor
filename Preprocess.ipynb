The source for this data is a point download near Pitsburgh from https://maps.nrel.gov/nsrdb-viewer/

```
import pandas as pd
```

```
for year in range(2000, 2018):
  with open('1064057_40.45_-80.02_'+str(year)+'.csv', 'r') as fin:
      data = fin.read().splitlines(True)
  with open('HEADLESS_1064057_40.45_-80.02_'+str(year)+'.csv', 'w') as fout:
      fout.writelines(data[2:])
```

Now, we add a Date column by adding the Day and Month column in string format. Then we can group the rows by date and find daily averages of the data. The daily averages are required for it to work with other data frames.



```
for year in range(2000, 2018):
  temp = pd.read_csv('/content/HEADLESS_1064057_40.45_-80.02_'+str(year)+'.csv',low_memory=False)
  temp['Date']=temp.apply(lambda row: row.Day.astype(str)+row.Month.astype(str),axis=1)
  writeable = temp.groupby(['Date']).mean()
  writeable.to_csv('/content/HEADLESS_1064057_40.45_-80.02_'+str(year)+'.csv')
```

In the last line of the loop, the groupby object is written as a csv file. It is overwritting the original headless file so now we have a headless file with daily averages, that is, 365 rows per csv file.
